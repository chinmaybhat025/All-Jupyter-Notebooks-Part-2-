```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://www.imdb.com/search/title/?title_type=tv_series,tv_miniseries&num_votes=5000,&genres=action&sort=year,asc"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
title = soup.title
print(title)
```

```python
# Print out the text
text = soup.get_text()
##print(soup.text)
```

```python
soup.find_all('a')
```

```python
all_links = soup.find_all("a")
for link in all_links:
    print(link.get("href"))
```

```python
# Print the first 10 rows for sanity check
rows = soup.find_all('span')
print(rows[:10])
```

```python
for row in rows:
    row_td = row.find_all('span/strong')
print(row_td)
type(row_td)
```
