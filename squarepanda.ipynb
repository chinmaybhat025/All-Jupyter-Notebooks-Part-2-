#### Rearrange columns for readability

```python
df=df[['id', 'childid', 'userid', 'gamename', 'nonsensewords', 'word',
       'wordtype', 'createdat', 'starttime', 'endtime', 'start_ms_time',
       'end_ms_time', 'duration', 'platform', 'dnc', 'gender', 'dateofbirth',
       'gamescore']]
```

#### Using only rows where dnc = 0.

```python
df1.nunique()
```

#### check Null values

```python
df1.isnull().sum()
```

#### Drop the rows(activity) with has no words or gamescore or start time from analysis

```python
df1 = df1.dropna(axis=0, subset=['word','gamescore','starttime','start_ms_time'])
```

#### Replaceing Other Null values which are not recorded 

```python
df1['platform'].replace(np.NaN,"Others",inplace=True)
df1['gender'].replace(np.NaN,"not recorded",inplace=True)
df1['dateofbirth'].replace(np.NaN,"not recorded",inplace=True)
df1.isnull().sum()
```

```python
df1.shape
```

#### Breaking Date time into components

```python
# day
df1['Start_Day'] = df1["createdat"].dt.day
# month
df1['Month'] = df1["createdat"].dt.strftime("%B")
# year
df1['Year'] = df1["createdat"].dt.year
# hour
df1['Start_hour'] = df1["createdat"].dt.hour
# minute
df1['Start_minute'] = df1["createdat"].dt.minute
# second
df1['Start_second'] =df1["createdat"].dt.second
# weekday
df1['Start_weekday'] =df1["createdat"].dt.strftime("%A")
```

```python
df1=df1[['id', 'childid', 'userid', 'gamename', 'nonsensewords', 'word',
       'wordtype','Start_Day','Start_weekday', 'Month', 'Year', 'Start_hour','Start_minute',
       'Start_second','duration', 'platform', 'dnc', 'gender', 'dateofbirth','gamescore']]
```

```python
df1=df1.reset_index()
```

```python
df1=df1.drop("index",axis=1)
```

# Questions

#### Q1 a)  ‘Session-ize’ the activities: Can you group activities in a way to create sessions out of activities that are close together? 

#### I have decided to create 3 sessions per day as morning_sessions,afternoon_sessions,night_sessions

```python
df_mrg=df1[df1["Start_hour"]<12]
```

```python
df_mrg.shape
```

```python
df_afternoon=df1[(df1["Start_hour"]>=12) & (df1["Start_hour"]<=18)]
```

```python
df_afternoon.shape
```

```python
df_night=df1[df1["Start_hour"]>18]
```

```python
df_night.shape
```

### Creating Session

#### Per day Morning sessions-Hour wise

```python
morning_sessions=df_mrg.groupby(["Start_Day","Start_hour"])
morning_sessions.first()
```

#### Per day afternoon sessions-Hour wise

```python
afternoon_sessions=df_afternoon.groupby(["Start_Day","Start_hour"])
afternoon_sessions.first()
```

#### Per day night sessions-Hour wise

```python
night_sessions=df_night.groupby(["Start_Day","Start_hour"])
night_sessions.first()
```

```python
df_afternoon.groupby(["Start_Day"])["id"].count().mean()
```

```python
df_night.groupby(["Start_Day"])["id"].count().mean()
```
