## Questions:
Sessions: As you can see, children can have multiple activities on the same day.
Some activities are close in time; while some are further apart.
- a. â€˜Session-izeâ€™ the activities: Can you group activities in a way to create
  sessions out of activities that are close together?
- b. How many activities per session on average (Over time)?
- c. How many different games do children play per session?

```python
df_mrg=df[df["Start_hour"]<12]
df_afternoon=df[(df["Start_hour"]>=12) & (df["Start_hour"]<=18)]
df_night=df[df["Start_hour"]>18]
```

```python
morning_sessions=df_mrg.groupby(["Start_Day","Start_hour"])
morning_sessions.first()
```

```python
afternoon_sessions=df_afternoon.groupby(["Start_Day","Start_hour"])
afternoon_sessions.first()
```

```python
night_sessions=df_night.groupby(["Start_Day","Start_hour"])
night_sessions.first()
```
